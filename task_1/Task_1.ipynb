```python
#расчет параметров precision, recall, F1-score
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, y_pred, average='macro')
```

### Задание 4. Кросс-валидация

```python
from sklearn.model_selection import StratifiedKFold
k_choices = (1, 2, 3, 5, 8, 10, 15, 20, 25, 50)
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits)
```

```python
def get_accuracy(k, indexes):
    train_index, validation_index = indexes
    part_train_X, part_train_y = train_X[train_index], train_y[train_index]
    validation_X, validation_y = train_X[validation_index], train_y[validation_index]
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(part_train_X, part_train_y)
    return knn.score(validation_X, validation_y)

accuracy_list = []

for k in k_choices:
    get_accuracy_k = lambda indexes: get_accuracy(k, indexes)
    sum_accuracy = sum(map(get_accuracy_k, skf.split(train_X, train_y)))
    accuracy_list.append((sum_accuracy / n_splits, k))
```

```python
for pair in accuracy_list:
    print('({:.2f}, {})'.format(*pair), end=' ')
best_pair = max(accuracy_list)
k = best_pair[1]
print('\nЛучшая точность {:.2f} на итерации с k = {}'.format(*best_pair))
```

```python
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(train_X, train_y)
```

```python
print('Accuracy: {}'.format(knn.score(test_X, test_y)))
```
